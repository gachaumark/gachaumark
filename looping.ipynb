LISTS AND LOOPS IN PYTHON


```python
# This line initializes a list named 'fruits' with three string elements: "mangoes", "Apples", and "Bananas".
fruits=["mangoes","Apples","Bananas"]
# This line starts a 'for' loop. It will iterate over each element in the 'fruits' list.
# In each iteration, the current fruit will be assigned to the variable 'y'.
for y in fruits :
  # This line prints the current value of 'y' (which is one of the fruits) to the console.
  print(y)
```

```python
languages="Heloo World"
for x in languages:
  print(x)
```

```python
numbers=[1,4,5,5,5,3,7,8,8,9,5,3,5,5,5,5,4,2]

# To fix this, you need to iterate through the list and check each number individually.
# For example, to print if each number is even or odd:
for num in numbers:
  if (num % 2) == 0:
    print(f"{num} is even")
  else:
    print(f"{num} is odd")
```

FINDING THE SUM OF A LIST

```python
nums=[23,43,56,34,21,12,54,45,3,27,78]
sum=0
for x in nums:
  sum+=x
  print(sum)
```

FINDING THE SUM OF ALL EVEN NUMBERS

```python
nums=[23,43,56,86,73,80,34,21,12,54,45,3,27,78]
for x in nums:
  sum=0
  if (x%2)==0:
    sum+=x
    print(f"THE TOTAL EVEN NUMBERS = {sum}")
```
