```python
#Enter the personal info
PayRollnumber=int(input("Enter the PayRoll Number:"))
Name=input("Enter the Name:")
Gender=input("Enter your Gender")
JobGroup=(input("Enter the JobGroup:"))
Salary=int(input("Enter the Basic Salary:"))
print("PayRoll Number :",PayRollnumber)
print("Your Name is :",Name)
if Gender=="male":
  print("You are Male")
elif Gender=="female":
  print("You are Female")
else:
  print("You GAYY OR a ladyBOYY!!")
print("Your JobGroup :",JobGroup)
print("Total Monthly Salary :",Salary)
#Decide the JobGroup
if JobGroup=='J':
  allowance=5000
elif JobGroup=="K":
  allowance=5500
elif JobGroup=="L":
  allowance=6000
elif JobGroup=="M":
  allowance=4000
else:
  print("Enter the correct JobGroup")
#Compute the salary,GrossSalary and Taxes
Grosspay=Salary+allowance
print("Grosspay:",Grosspay)
Taxes=12*Salary/100
print("Taxes:",Taxes)
Netpay=Grosspay-Taxes
print("Netpay:",Netpay)


```
